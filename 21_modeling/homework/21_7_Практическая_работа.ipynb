```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt

df = pd.read_csv('data/hw1.csv')
df.head()
```

```python
plt.hist(df["X"]);
```

```python
plt.hist(df["Y"]);
```

```python
# Ваш код здесь
t_stat, p_value = stats.ttest_ind(df["X"], df["Y"])
print("T_stat:", t_stat, "p_value:", p_value)
if p_value < 0.05: print("H0 отвергается")
else: print("H0 принимается")
```

```python
df2 = pd.read_csv('data/hw2.csv')
df2.head()
```

```python
plt.hist(df2["before"]);
```

```python
plt.hist(df2["after"]);
```

```python
# Ваш код здесь
t_stat, p_value = stats.ttest_rel(df2.before, df2.after, alternative='greater')
print("T_stat:", t_stat, "p_value:", p_value)
if p_value < 0.05: print("H0 отвергается")
else: print("H0 принимается")
```

```python
df3 = pd.read_csv('data/hw3.csv')
df3.head()
```

```python
# Ваш код здесь
t_stat, p_value = stats.wilcoxon(df3.before,df3.after, alternative='less') 
print("T_stat:", t_stat, "p_value:", p_value)
if p_value < 0.05: print("H0 отвергается")
else: print("H0 принимается")
```
